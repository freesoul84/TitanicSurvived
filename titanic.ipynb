#### conversion of string to dummy variables

```python
pd.get_dummies(dataset['Sex'])
```

```python
sex=pd.get_dummies(dataset['Sex'],drop_first=True)
```

```python
sex.head(4)
```

```python
embarked=pd.get_dummies(dataset['Embarked'],drop_first=True)
embarked.head(6)
```

```python
pclass=pd.get_dummies(dataset['Pclass'],drop_first=True)
pclass.head(5)
```

#### adding these new columns into original dataset

```python
dataset.drop(['Q','S'],axis=1,inplace=True)
dataset
```

```python
dataset=pd.concat([dataset,sex,embarked,pclass],axis=1)
```

```python
dataset
```

#### model use is logistic regression

## training data

```python
X=dataset.drop('Survived',axis=1)
Y=dataset['Survived']
```

```python
X
```

```python
Y
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_predict=classifier.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import classification_report
classification_report=classification_report(Y_test,Y_predict)
classification_report
```

### to check wrong predictions

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,Y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_predict)
```
